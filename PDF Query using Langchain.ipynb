```python
pdfreader=PdfReader('PDS.pdf')
```

```python
from typing_extensions import Concatenate
raw_text=''
for i, page in enumerate(pdfreader.pages):
  content=page.extract_text()
  if content:
    raw_text+=content
```

```python
raw_text
```

```python
text_splitter=CharacterTextSplitter(separator="\n",
                                    chunk_size=1000
                                    ,chunk_overlap=200
                                    ,length_function=len,)

texts=text_splitter.split_text(raw_text)
```

```python
len(texts)
```

```python
sub_text=texts[1:50]
```

```python
embeddings=OpenAIEmbeddings()
```

```python
document_search=FAISS.from_texts(sub_text, embeddings)
```

```python
document_search
```

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
```

```python
chain=load_qa_chain(OpenAI(), chain_type="stuff")
```

```python
query1="What is the flat adminstration fee? Give me only the dollar value"
docs=document_search.similarity_search(query1)
sol1=chain.run(input_documents=docs, question=query1)
```

```python
query2="What is the asset based fee? Give me only the cost of this fee"
docs=document_search.similarity_search(query2)
sol2=chain.run(input_documents=docs, question=query2)
```

```python
query3="What is the cap on the asset based fee?"
docs=document_search.similarity_search(query3)
sol3=chain.run(input_documents=docs, question=query3)
```

```python
flat_admin_fee=re.findall('[\$\d+]+\D+', sol1)
asset_based_fee=re.findall('\d+.*', sol2)
asset_based_cap=re.findall('\$\d+.*', sol3)

# sol1.substr(sol1.indexOf("$") + 1);
```
